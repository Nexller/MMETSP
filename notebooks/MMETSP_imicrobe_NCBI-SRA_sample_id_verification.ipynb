# MMETSP samples on imicrobe vs. NCGR

The SRA BioProject id for the MMETSP is: [PRJNA231566](https://www.ncbi.nlm.nih.gov/bioproject/?term=PRJNA231566)
This contains 678 samples (MMETSP id in SampleName) and 719 SRA Experiments (Run).
Note that there are 2 sub-projects, [PRJNA248394](https://www.ncbi.nlm.nih.gov/bioproject/?term=PRJNA231566) with 718 samples and [PRJNA245816](https://www.ncbi.nlm.nih.gov/bioproject/245816) with 1 extra Run from Tetraselmis sp. GSL018.

Files from [imicrobe](https://imicrobe.us/?#/projects/104) were downloaded from the [ftp site](http://bit.ly/2frEwwl).

```r
imicrobe<-read.csv("../imicrobe/Callum_FINAL_biosample_ids.csv")
ncbi<-read.csv("../SraRunInfo.csv")
ncbi_719<-read.csv("../SraRunInfo_719.csv")
source('../extra_scripts/overLapper.R') # plots venn diagrams
MMETSP_id_NCBI<-ncbi$SampleName
MMETSP_id_imicrobe<-imicrobe$SAMPLE_NAME
```

```r
setlist <- list(NCBI=as.vector(MMETSP_id_NCBI),imicrobe=as.vector(MMETSP_id_imicrobe))
OLlist <- overLapper(setlist=setlist, sep="", type="vennsets")
counts <- sapply(OLlist$Venn_List, length)
vennPlot(counts=counts)
```

## Overlap

```r
overlap<-intersect(MMETSP_id_NCBI,MMETSP_id_imicrobe)
length(overlap)
```
