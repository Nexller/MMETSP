```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
evaluation_frame = pd.read_csv("../assembly_evaluation_data/MMETSP_evaluation_matrix.csv",index_col=0)
evaluation_frame.head()
list(evaluation_frame.columns.values)
```

```python
reads_frame = evaluation_frame[['Input.Reads', 'Unique_kmers','n_seqs.x','score','Complete_BUSCO_perc','p_refs_with_CRBB','mean_orf_percent.y','linguistic_complexity.y','Phylum']]
```

```python
reads_frame.head()
```

```python
g = sns.PairGrid(reads_frame)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
```

```python
g = sns.PairGrid(reads_frame, hue="Phylum")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```
